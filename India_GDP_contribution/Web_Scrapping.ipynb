```python
URL="https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_GDP"
print("URL is copied")
```

```python
tables=pd.read_html(URL)
print(f"total tables found:{tables}")
```

```python
#Iterating through each table to find the correct index
for idx, table in enumerate(tables):
    print("\nTable{idx}")
    print(table.head())
```

```python
#Extraction of the Tables

df=tables[1]
#Replace column headers with column numbers
df.columns=range(df.shape[1])
df=df[[0,1]]
df=df.iloc[0:34,:]
df.columns=["State/Union_Terrirtory", "2022-23(in rupees billions)"]
df.head(34)
```

```python
df.to_csv('./India_GDP_COntribution.csv')
```
