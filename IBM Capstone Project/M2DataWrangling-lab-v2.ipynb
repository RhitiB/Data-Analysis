### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_counts=df.isnull().sum().idxmax()
print(missing_counts)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
df["ConvertedCompYearly"].isnull().sum()
df["ConvertedCompYearly"].fillna(df["ConvertedCompYearly"].median(), inplace=True)
```

```python
df["ConvertedCompYearly"].isnull().any()
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
df["RemoteWork"].isnull().any()
df["RemoteWork"].fillna(df["RemoteWork"].mode()[0], inplace=True)
```

```python
df["RemoteWork"].isnull().any()
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
df["ConvertedCompYearly_normalized"]=(df["ConvertedCompYearly"]-df["ConvertedCompYearly"].min())/(df["ConvertedCompYearly"].max()-df["ConvertedCompYearly"].min())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np
df['ConvertedCompyYearly'] = np.log(df['ConvertedCompYearly'])
df.head()
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
def convert_experience(value):
    if isinstance(value, str):
        if "less" in value.lower():  # Handle "Less than 1 year"
            return 0
        elif "more" in value.lower():  # Handle "More than 50 years"
            return 50
    return pd.to_numeric(value, errors='coerce')  # Convert others to numeric

df['YearsCodePro'] = df['YearsCodePro'].apply(convert_experience)

# Define Experience Levels using pd.cut
df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], 
                               bins=[-1, 0, 2, 6, 14, float('inf')], 
                               labels=['Beginner', 'Junior', 'Mid-Level', 'Senior', 'Expert'])

(df.head())
```
